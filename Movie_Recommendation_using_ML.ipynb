# **Objective**



---


The objective of the movie recommendation system using machine learning based on vectorization and cosine similarities is to provide personalized and accurate movie recommendations to users based on their preferences and similarities to other users. The system aims to leverage machine learning techniques to analyze movie features and user behavior in order to create a recommendation model that effectively matches users with movies they are likely to enjoy.

**Recommendation Accuracy:** Develop a robust machine learning model that accurately predicts the preferences of users based on their historical movie ratings and other relevant features. The system should strive to minimize errors and improve the precision and recall of the recommendations.

**Vectorization and Cosine Similarities:**Utilize vectorization techniques to represent movies and users in a high-dimensional space. Calculate cosine similarities between these vectors to measure the similarity between movies and users. The objective is to effectively capture the relationships between movies and users, enabling accurate recommendation generation.

By achieving these objectives, the movie recommendation system aims to enhance user satisfaction, increase movie discovery, and improve engagement, ultimately driving increased user retention and revenue for the platform.

# **Modeling and Data Preprocessing**

```python
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix= tfidf.fit_transform(MR['overview'])
tfidf_matrix.shape
```

```python
cosine_sim= linear_kernel(tfidf_matrix,tfidf_matrix)
cosine_sim
```

```python
indices = pd.Series(MR.index, index=MR['title_x'])
type(indices)
indices
```

# **Data Segmentation**

```python
def recommend(title_x,cosine_sim=cosine_sim):
  idx = indices[title_x]
  sim_scores = list(enumerate(cosine_sim[idx]))
  sim_scores = sorted(sim_scores, key=lambda a: a[1], reverse=True)
  sim_scores=sim_scores[1:11]
  movie_indices = [i[0] for i in sim_scores]
  print(type(MR['title_x'].iloc[movie_indices]))
  return pd.DataFrame(MR['title_x'].iloc[movie_indices])
```

# **Prediction or Output**

```python
recommend('Iron Man')
```

# **Rating vs Votes**

```python
vote_counts = MR[MR['vote_count'].notnull()]['vote_count'].astype('int')
vote_averages = MR[MR['vote_average'].notnull()]['vote_average'].astype('int')
MEAN=vote_averages.mean()
MEAN
```

```python
Mean_ac=vote_counts.quantile(0.99)
Mean_ac
```

```python
def weighted_rating(x):
  v=x['vote_count']
  R=x['vote_average']
  return (v/(v+Mean_ac)*R)+(Mean_ac/(Mean_ac+v)*MEAN)
```

```python
MR['rating_count']=MR.apply(weighted_rating,axis=1)
MR.head()
```
